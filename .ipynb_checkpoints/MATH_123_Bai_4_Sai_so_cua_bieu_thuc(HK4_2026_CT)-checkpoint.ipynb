###  Sai số tương đối :

## Các ví dụ : 1)

###  Code sai số của biểu thức :

```python
import sympy as sp
x, y = sp.symbols('x,y')
f = x**3+x**2*y+y**3
print('Biểu thức f =')
display(f)
```

```python
x0 = 2.5342
print('Giá trị x0 = ')
display(x0)
saiso_x0 = 0.0032
print('Sai số tuyệt đối của x0 :')
display(saiso_x0)
y0 = 2.6578
print('Giá trị y0 = ')
display(y0)
saiso_y0 = 0.0065
print('Sai số tuyệt đối của y0 :')
display(saiso_y0)
```

```python
df_dx = sp.diff(f, x)
print('Đạo hàm riêng theo x :')
display(df_dx)
df_dy = sp.diff(f, y)
print('Đạo hàm riêng theo y :')
display(df_dy)
```

```python
a1 = sp.Abs(df_dx.subs([(x,x0),(y,y0)]))* saiso_x0
a2 = sp.Abs(df_dy.subs([(x,x0),(y,y0)]))* saiso_y0
```

```python
saiso_bt = a1 + a2
print('Sai số tuyệt đối của biểu thức tại (x0,y0) là :')
display(saiso_bt)
```

```python
print('Giá trị của biểu thức tại (x0,y0) là :')
gt = f.subs([(x,x0),(y,y0)])
display(gt)
print('Sai số tương đối của biểu thức tại (x0,y0) là (theo %):')
display(saiso_bt/abs(gt)*100)
```

### Ghép các đoạn code trên lại với nhau, bổ bớt các phần in ra trên màn hình không cần thiết, sao cho chỉ thể hiện các kết quả chính như sau :

## Trình bày bài khi thi  :

```python
import sympy as sp
x, y = sp.symbols('x,y')
f = x**3+x**2*y+y**3
x0 = 2.5342
saiso_x0 = 0.0032
y0 = 2.6578
saiso_y0 = 0.0065
df_dx = sp.diff(f, x)
print('Đạo hàm riêng theo x :')
display(df_dx)
df_dy = sp.diff(f, y)
print('Đạo hàm riêng theo y :')
display(df_dy)
a1 = sp.Abs(df_dx.subs([(x,x0),(y,y0)]))* saiso_x0
a2 = sp.Abs(df_dy.subs([(x,x0),(y,y0)]))* saiso_y0
saiso_bt = a1 + a2
print('Sai số tuyệt đối của biểu thức tại (x0,y0) là :')
display(saiso_bt)
print('Giá trị của biểu thức tại (x0,y0) là :')
gt = f.subs([(x,x0),(y,y0)])
display(gt)
print('Sai số tương đối của biểu thức tại (x0,y0) là (theo %):')
display(saiso_bt/abs(gt)*100)
```
