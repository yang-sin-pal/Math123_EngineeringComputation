## 1) Đạo hàm của hàm số 1 biến :

### Tính đạo hàm của hàm số :

```python
import sympy as sp
x = sp.symbols('x')  
f =  x**2 + 2*x+1
f_dh = sp.diff(f,x)
display(f_dh)
```

###  Tính đạo hàm của hàm sau :

```python
import sympy as sp
x = sp.symbols('x')  
f =  (sp.E**(2*x))/(x*sp.cos(x))
f_dh = sp.diff(f,x)
display(f_dh)
```

### Đạo hàm cấp 2 của hàm :

```python
import sympy as sp
x = sp.symbols('x')  
f =  sp.exp(2*x)/(x*sp.cos(x))
f_dh2 = sp.diff(f,x,2)
display(f_dh2)
```

## 2)   Lệnh rút gọn :

```python
import sympy as sp
x = sp.symbols("x" , real=True) 
f = sp.sin(x)**2+sp.cos(x)**2
display(f)
```

```python
g = sp.simplify(f)
display(g)
```

###  Tính đạo hàm của hàm số sau và rút gọn kết quả :
### $$\displaystyle \frac{x \sqrt{x^{2} + 5}}{2} + \frac{5 \ln{\left(x + \sqrt{x^{2} + 5} \right)}}{2}$$

```python
import sympy as sp
x= sp.symbols('x' , real=True)  

f = ((x*sp.sqrt((x**2)+5))/2)+((5*sp.ln(x+sp.sqrt((x**2)+5))/2))
print('f:')
display(f) 

f_dh = sp.diff(f,x)
print("f_dh:")
display(f_dh)

g = sp.simplify(f_dh)
print("g = f_dh simplified:")
display(g)

```

###  Hàm nhiều biến và đạo hàm riêng 

```python
import sympy as sp
x,y = sp.symbols('x,y')    
f = sp.exp(-3*x)*sp.sin(2*y)
display(f)
```

```python
dhf_x = sp.diff(f,x)
display(dhf_x)
```

```python
dhf_y = sp.diff(f,y)
display(dhf_y)
```

###  Nghịch đảo của ma trận vuông A là một ma trận cùng kích thước với A, sao cho khi nhân với A   ( phía trước và phía sau ) đều cho ra ma trận đơn vị  (I)
## $A{A^{ - 1}}= I$
## ${A^{ - 1}}A= I$

## $$A=\displaystyle \left[\begin{matrix}1 & 2\\1 & 3\end{matrix}\right]$$

## ${A^{ - 1}}=\displaystyle \left[\begin{matrix}3 & -2\\-1 & 1\end{matrix}\right]$

```python
import sympy as sp
K=[3,-2,-1,1]
A=sp.Matrix(2,2,K)
print('Ma trận A là: ')
display(A)
print('Ma trận nghịch đảo của A là: ')
display(A**(-1))   
```

```python
import sympy as sp
K=[1,-2,3,4,5,6,7,8,-9]
A=sp.Matrix(3,3,K)
display(A)
print('Chuẩn của ma trận A là :')
q = A.norm(sp.oo)
display(q)
```

##  4)  Công thức Taylor

### Áp dụng cho hàm sin(x) và với n=5 :

```python
import sympy as sp
x = sp.symbols('x', real=True )
f = sp.sin(x)
n = 5  
x0 = 0  
f_taylor = f.subs(x, x0)
for i in range(1, n + 1, 1):
    dh_i = sp.diff(f, x, i)
    dh_i_x0 = sp.simplify(dh_i).subs(x, x0)
    print('Đạo hàm cấp',i,'của hàm số tại x=',x0, 'là:')
    display(dh_i_x0)
    f_taylor = f_taylor + dh_i_x0 * (x - x0)**i /sp.factorial(i)
print('Áp dụng vào công thức ta được khai triển Taylor của hàm tại x0=',x0,'là:')
display(f_taylor)
```
