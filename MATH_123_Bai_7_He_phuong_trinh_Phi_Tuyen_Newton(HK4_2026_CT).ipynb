### Bước 1 : Nhập dữ liệu

```python
import sympy as sp
x, y = sp.symbols('x ,y')
f = x**2 + y**2 - 1
g = x **3 -y     
x0 = 0.9  
y0 = 0.5
n = 10
x_nghiem = [x0]
y_nghiem = [y0]
```

### Bước 2 : Tính giá trị bước tiếp theo

```python
for i in range(1, n + 1,1): 
    f0 = f.subs([(x,x0),(y,y0)])
    g0 = g.subs([(x,x0),(y,y0)])
    fx = sp.diff(f,x).subs([(x,x0),(y,y0)])
    fy = sp.diff(f,y).subs([(x,x0),(y,y0)])
    gx = sp.diff(g,x).subs([(x,x0),(y,y0)])
    gy = sp.diff(g,y).subs([(x,x0),(y,y0)])
    TSx = sp.det(sp.Matrix(2,2,[-f0,fy,-g0,gy] ))
    TSy = sp.det(sp.Matrix(2,2,[-fx,f0,-gx,g0] ))
    MS =  sp.det(sp.Matrix(2,2,[fx,fy,gx,gy] ))
    x1 = x0 + TSx/MS
    y1 = y0 + TSy/MS
    x_nghiem.append(x1)
    y_nghiem.append(y1)
    print( 'Giá trị x bước lặp thứ',i,' là :')
    display(x_nghiem[i])
    print( 'Giá trị y bước lặp thứ',i,' là :')
    display(y_nghiem[i])
    x0 = x_nghiem[i]
    y0 = y_nghiem[i]   
```
