```python
a = 0.0   ### Chú ý : Không nhập 0 mà nhập 0.0 
b = 1.0     
n = 5
nghiem = [a]   
```

```python
for i in range(1, n + 1 , 1):
    c = (a + b) / 2
    nghiem.append(c)
    if f.subs(x,c) * f.subs(x,a) < 0:
      b = c
    else:
      a = c
    print('Nghiệm gần đúng tại bước chia đôi thứ',i,'là:',nghiem[i])
```

```python
print('Sai số của phương pháp chia đôi tại bước lặp thứ ',n, 'là :')
display(sp.Abs(b-a)/2**n)
```

### 📝 Giải thích hiện tượng kết quả "đứng im" từ bước 54:

* **Nguyên nhân:** Do giới hạn **độ chính xác số thực 64-bit (Float64)** của máy tính. Kiểu dữ liệu này chỉ biểu diễn chính xác tối đa **15 - 17 chữ số thập phân**.
* **Cơ chế lỗi:** Tại bước 54, khoảng cách giữa `a` và `b` nhỏ hơn giới hạn nhận biết của máy tính ($\Delta < \text{Machine Epsilon} \approx 2.22 \times 10^{-16}$).

  Phép tính `(a + b) / 2` bị làm tròn về lại giá trị cũ, khiến nghiệm không thể chia nhỏ hơn được nữa.
* **Giải pháp:** Nếu muốn độ chính xác cao hơn, cần khai báo biến ban đầu bằng kiểu dữ liệu có độ chính xác tùy ý của SymPy:

  `a = sp.Float(0.0, 30)` và `b = sp.Float(1.0, 30)`

```python
import sympy as sp
x = sp.symbols('x')
f = x-sp.cos(x)
a = sp.Float(0.0, 30) ## Áp dụng giải pháp
b = sp.Float(1.0, 30) ## Áp dụng giải pháp
d = b-a
n = 60  
nghiem = [a]
for i in range(1, n + 1, 1):
    c = (a + b) / 2
    nghiem.append(c)
    if f.subs(x,c) * f.subs(x,a) < 0:
      b = c
    else:
      a = c
    if i >= 50:
        print('Nghiệm gần đúng tại bước chia đôi thứ',i,'là:',nghiem[i])
```
