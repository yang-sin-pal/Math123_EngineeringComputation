### Viết code để  in ra màn hình các hàm sau đây:

```python
import sympy as sp
from IPython.display import Math

x = sp.symbols('x')

# 1. Dùng sp.Add(..., evaluate=False) để ép giữ nguyên thứ tự tan(x) + log(x)
bieu_thuc_y = sp.Add(sp.tan(x), sp.log(x), evaluate=False)
y = sp.sqrt(bieu_thuc_y)

# 2. Dùng sp.Add(..., evaluate=False) để ép giữ nguyên thứ tự x^2 + 1
bieu_thuc_a = sp.Add(x**2, 1, evaluate=False)
a = sp.root(bieu_thuc_a**2, 3)

# Biến đổi biểu thức SymPy sang định dạng LaTeX rồi nối với nhau bằng dấu phẩy
# Ép latex() đổi thành 'ln' thay vì 'log' (ln_notation=True)
chuoi_y = sp.latex(y, ln_notation=True)
chuoi_a = sp.latex(a, ln_notation=True)

# Nối chuỗi lại với nhau
chuoi_hien_thi = chuoi_y + r' \quad , \quad ' + chuoi_a

# Hiển thị chuỗi toán học đẹp mắt không chứa dấu ngoặc của bộ (tuple)
display(Math(chuoi_hien_thi))
```

### Viết code để in ra màn hình hàm sau đây:

## $$\displaystyle \frac{x^{2}}{2 \sqrt{x^{2} + 5}} + \frac{\sqrt{x^{2} + 5}}{2} + \frac{5 \left(\frac{x}{\sqrt{x^{2} + 5}} + 1\right)}{2 \left(x + \sqrt{x^{2} + 5}\right)}$$

```python
import sympy as sp
x = sp.symbols('x')
a = sp.sqrt(x**2 + 5)
b = (x**2)/(2*a)
c = a/2

# d = (5*((x/a) + 1))/(2*(x + a))
# Giữ nguyên 5 * (...) và 2 * (...) không cho nhân vào trong
tu_so = sp.Mul(5, (x/a) + 1, evaluate=False)
mau_so = sp.Mul(2, x + a, evaluate=False)
d = tu_so / mau_so

y= b + c + d
print('Hàm số y=')
display(y)    
```

### 7. Nhập hàm và in ra giá trị của hàm số tại 1 điểm cho trước

```python
import sympy as sp

x = sp.symbols("x") 
y = sp.sin(x)**3
x0 = sp.pi/6
print('Hàm số y=')
display(y)

print('Giá trị hàm số y tại điểm đã cho là:')
y0 = y.subs(x,x0)     
display(y0)
```

```python
import sympy as sp
from IPython.display import Math

# 1. Khai báo ký hiệu và hàm số
x = sp.symbols("x")
y = sp.sin(x)**3
x0 = sp.pi/6

# Sử dụng Math để trộn chữ tiếng Việt và LaTeX trên cùng 1 dòng
display(Math(rf"\text{{Hàm số }} y = {sp.latex(y)}"))

print('---')
print('Biểu thức tính toán:')

# 2. Tạo vế trái: Giữ nguyên sin(pi/6)^3 không cho tự động tính toán
vế_trái = sp.sin(sp.UnevaluatedExpr(x0))**3

# 3. Tạo vế phải: Dùng sp.Subs để tạo ký hiệu thế số dọc "|"
vế_phải = sp.Subs(y, x, x0)

# 4. Gom lại thành một phương trình bằng sp.Eq và hiển thị
biểu_thức_mẫu = sp.Eq(vế_trái, vế_phải)
display(biểu_thức_mẫu)

print('---')
print('Giá trị của hàm số bằng:')
# 5. Tính toán giá trị thực tế
y0 = y.subs(x, x0)
display(y0)
```
