```python
from IPython.display import display, Math
import sympy as sp

x = sp.symbols('x')
phi = 1000.0 - x ** 3
a = 9.0
b = 10.0
x0 = 10.0
n = 8

x_gd = [x0]
for i in range(1, n + 1, 1):
    x_sau = phi.subs(x, x_gd[i - 1])
    x_gd.append(x_sau)
    
    s = f"{float(x_gd[i]):.14e}"
    if "e" in s:
        base, exponent = s.split("e")
        exponent = exponent.replace("+", "")
        if int(exponent) != 0:
            # Dạng mã LaTeX để hiển thị dấu chấm nhân và số mũ trên đầu
            formatted_num = f"{base} \\cdot 10^{{{exponent}}}"
        else:
            formatted_num = base
    else:
        formatted_num = s
        
    display(Math(f'x_{{\\text{{gd}}}}[ {i} ] \\text{{ là: }} {formatted_num}'))
```

### b) Trường hợp  q<1  ( dùng được )

```python
import sympy as sp
x = sp.symbols('x')
phi = (1000-x)**(sp.Rational(1,3))
a = 9.0
b = 10.0
x0 = 10.0 
n = 8
x_gd = [x0] 
for i in range(1, n+1 , 1): 
    x_sau = phi.subs(x,x_gd[i-1])
    x_gd.append(x_sau)
    print('x_gd[',i,'] là:',x_gd[i])
```

### Tính sai  số của nghiệm $x_4$

```python
print('Đạo hàm của hàm phi(x) là :')
phi_dh = sp.diff(phi, x)
display(phi_dh)
M = sp.N(sp.maximum(phi_dh,x,sp.Interval(a,b)))
m = sp.N(sp.minimum(phi_dh,x,sp.Interval(a,b)))
q = max(abs(M),abs(m))
print('Giá trị của hệ số q là :')
display(q)
k = 4
sai_so  = abs(x_gd[k] - x_gd[k-1]) * q / (1 - q)
print('Sai số của nghiệm x_gd[',k,'] là :')
display(sai_so)
```
